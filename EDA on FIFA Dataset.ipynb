```python
#Loading the required libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#Loading the required dataset
fifa=pd.read_csv("players_20.csv")
```

```python
#Having a glance at first 05 rows of dataset
fifa.head()
```

```python
#Extracting all the columns of the dataset
for col in fifa.columns:
  print(col)
```

```python
#Looking at the shape of dataframe
fifa.shape
```

```python
#How many total entries are there in the dataframe
fifa.shape[0]*fifa.shape[1]
```

```python
#Finding out the number of players from each nation
fifa["nationality"].value_counts()
```

```python
#Finding out the top 10 nations with the most number of players
fifa["nationality"].value_counts()[0:10]
```

```python
#Make a bar plot for top 5 nations with the most number of players
plt.bar(fifa["nationality"].value_counts()[0:5].keys(),fifa["nationality"].value_counts()[0:5])
plt.title("Top 05 Nations With Most Number Of Players")
plt.xlabel("Name Of Country")
plt.ylabel("Player Count")
plt.show()
```

```python
sns.barplot(x=fifa["nationality"].value_counts()[0:5].keys(),y=fifa["nationality"].value_counts()[0:5])
plt.show()
```

```python
#Extract the columns from the dataset containing information about player names and their respective wages
player_salary=fifa[["short_name","wage_eur"]]
```

```python
#Looking at the head of above dataset
player_salary.head()
```

```python
#Sort the above dataset based on the wages
player_salary=player_salary.sort_values(by="wage_eur",ascending=False)
```

```python
#Make a Bar Plot for the top 5 players with maximum wages
plt.bar(player_salary["short_name"][0:5],player_salary["wage_eur"][0:5])
plt.title("Bar Plot For Players With Maximum Wages")
plt.xlabel("Player Name")
plt.ylabel("Wages in Euro")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.barplot(x=player_salary["short_name"][0:5],y=player_salary["wage_eur"][0:5])
plt.xticks(rotation=90)
plt.show()
```
